```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df_final_demo is your DataFrame

# Create the histogram with transparency and layered or dodged bars
sns.histplot(data=df_final_demo, x='clnt_age', hue='gendr', multiple='layer', alpha=0.15)  # or multiple='dodge'

# Customize the plot (optional)
plt.xlabel('Client Age')
plt.ylabel('Frequency')
plt.title('Histogram of Client Age by Gender')

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df_final_demo is your DataFrame and 'gendr' has four unique values

# Create a FacetGrid
g = sns.FacetGrid(df_final_demo, col='gendr', col_wrap=2)  # col_wrap=2 for 2 columns

# Map the histogram to each subplot
g.map(sns.histplot, 'clnt_age')

# Customize the plot (optional)
g.set_axis_labels('Client Age', 'Frequency')
g.fig.suptitle('Histograms of Client Age by Gender', y=1.02)  # Adjust y for title position

# Display the plot
plt.show()
```

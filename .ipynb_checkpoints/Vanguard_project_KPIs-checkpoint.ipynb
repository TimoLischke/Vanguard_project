```python
# Timo: KPI "time spent on each step" grouped by visit_id:

# Convert 'date_time' to datetime format
df_final_web_data['date_time'] = pd.to_datetime(df_final_web_data['date_time'])

# Calculate time spent on each step (time difference with the previous row)
df_final_web_data['time_spent'] = df_final_web_data.groupby('visit_id')['date_time'].diff()

# Display the result
print(df_final_web_data)
```

```python
df_final_web_data.head(20)
```

```python
# Replace time_spent with 0 in datetime format where process_step == 0
df_final_web_data.loc[df_final_web_data['process_step'] == 0, 'time_spent'] = pd.Timedelta(0)
df_final_web_data.head(20)


# Would be recommended to also set all FIRST entries for process_step with a given visit_id to 0.
```

```python
# neue spalte time_spent_in_sec hinzufÃ¼gen
df_final_web_data['time_spent_seconds'] = df_final_web_data['time_spent'].dt.total_seconds()
df_final_web_data.head(20)
```

```python
df_final_web_data.head(100)
```

```python
df_final_web_data.sort_values(by='time_spent_seconds', ascending=False)
```

```python
# Plotting a box plot to visualize the 'time_spent' for each 'process_step', using the "coolwarm" color palette.
sns.boxplot(data=df_final_web_data, x='process_step', y='time_spent_seconds', palette="coolwarm")
plt.show()
```

```python
df_final_web_data[df_final_web_data['time_spent_seconds'] >5000]
```

```python
# Delete Outliers according to tukeyÂ´s method
# Compute Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df_final_web_data['time_spent_seconds'].quantile(0.25)
Q3 = df_final_web_data['time_spent_seconds'].quantile(0.75)

# Compute the Interquartile Range (IQR)
IQR = Q3 - Q1

# Define lower and upper bounds for non-outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the DataFrame to remove outliers
df_final_web_data_no_time_outliers = df_final_web_data[(df_final_web_data['time_spent_seconds'] >= lower_bound) & (df_final_web_data['time_spent_seconds'] <= upper_bound)]

# Display results
print("\nFiltered DataFrame (No Outliers):\n", df_final_web_data_no_time_outliers)
```

```python
# Plotting a box plot to visualize the 'time_spent' for each 'process_step' after removing the time_spent outliers.
sns.boxplot(data=df_final_web_data_no_time_outliers, x='process_step', y='time_spent_seconds', palette="coolwarm")
plt.xlabel("Process Step")
plt.ylabel("time spent (seconds)")
plt.title("Time spent per process step of Vanguard clients")

# Safe the plot (needs to be done first, before showing it!)
# plt.savefig("Vanguard_clients_time_spent_on_process_step.png", dpi=300)
####  ONLY RUN ONCE !!!! ####

plt.show()
```

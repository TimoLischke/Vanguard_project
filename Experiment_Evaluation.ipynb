```python
df_final_unique = df_final.drop_duplicates(subset=["client_id"])
df_final_unique
```

```python
crosstab_1 = pd.crosstab(df_final_unique['gendr'], df_final_unique['Variation'])
crosstab_1
```

```python
crosstab_2 = pd.crosstab(
    index=df_final_unique['gendr'],  # Use 'gender' as rows
    columns=df_final_unique['Variation'],  # Use 'Variation' as columns
    values=df_final_unique['clnt_age'],  # Compute the mean of 'clnt_age'
    aggfunc='mean'  # Aggregate using mean
).round(2)

crosstab_2
```

```python
crosstab_3 = pd.crosstab(
    index=df_final_unique['gendr'],  # Use 'gender' as rows
    columns=df_final_unique['Variation'],  # Use 'Variation' as columns
    values=df_final_unique['clnt_tenure_yr'],  # Compute the mean of 'clnt_tenure_yr'
    aggfunc='mean'  # Aggregate using mean
).round(2)

crosstab_3
```

```python
crosstab_4 = pd.crosstab(
    index=df_final['Variation'],  
    columns="Min Date",  
    values=df_final['date_time'],  
    aggfunc='min'
).join(
    pd.crosstab(
        index=df_final['Variation'],  
        columns="Max Date",  
        values=df_final['date_time'],  
        aggfunc='max'
    )
)
# Convert date columns to datetime format (if not already)
crosstab_4["Min Date"] = pd.to_datetime(crosstab_4["Min Date"])
crosstab_4["Max Date"] = pd.to_datetime(crosstab_4["Max Date"])

# Compute total duration as the difference between Max and Min dates
crosstab_4["Total Duration"] = crosstab_4["Max Date"] - crosstab_4["Min Date"]
crosstab_4
```
